2a) define assymptomatic notations

```python
#define big -o notation
def big_o_notation(n):
    if n == 0:
        return "O(1)"
    elif n == 1:
        return "O(n)"
    elif n == 2:
        return "O(n^2)"
    else:
        return "O(n^k)"  # where k is a constant greater than 2
```

```python
#ii. define Big-Θ notation
def big_theta_notation(n):
    if n == 0:
        return "Θ(1)"
    elif n == 1:
        return "Θ(n)"
    elif n == 2:
        return "Θ(n^2)"
    else:
        return "Θ(n^k)"  # where k is a constant greater than 2
```

```python
#ii. Big-Ω - this desribes the lower bond of an algorithmns growth rate 
def big_omega_notation(n):
    if n == 0:
        return "Ω(1)"
    elif n == 1:
        return "Ω(n)"
    elif n == 2:
        return "Ω(n^2)"
    else:
        return "Ω(n^k)"  # where k is a constant greater than 2
```

i. Big-O - this describes the upper bond of an  algorithm`s growth rate 
ii. Big-Θ - describes both the upper and lower bonds of an algorithmns growth rate meaning it represents the exact rate of the growth
iii. Big-Ω - this desribes the lower bond of an algorithmns growth rate

3)SEARCHING AND SORTING ALGORITHMS 

```python
arr=[21,10,45,32,67,5,18]
```

```python
#a)python fuction to implement linear search tofind the value 32
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
print(linear_search(arr, 32))
```

```python
#b) Python function to implement binary search to find the value 45    assuming the arr is sorted
def binary_search(arr, target):
    arr.sort() 
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return -1
print(binary_search(arr, 45))
```

```python
#c)implement the bubble sort in ascending order
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
print(bubble_sort(arr))
```

LINKED LIST IMPLEMETATIONS

```python
# a) define a node class in python for a singly linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    #each node contains data and a reference to the next node in the list
```

```python
#linked list class
class linkedlist:
    def __init__(self):
        self.head = None 
```

```python
#insert node at the beginning
def insert_at_beginning(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node
```

```python
#insert at the end
def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:   # If list is empty
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
```

```python
#display all elements
def display(self):
    if self.head is None:
        print("The linked list is empty.")
        return
    current = self.head
    while current:
        print(current.data, end=" -> ")
        current = current.next
    print("None")
```

```python
#serving the firt student in the queue
def serve_student():
    if len(queue) == 0:
        print("No students in the queue to serve.")
    else:
        served = queue.pop(0)
        print(f"{served} has been served.")
```

```python
#function to display the current queue
def display_queue():
    if len(queue) == 0:
        print("The queue is empty.")
    else:
       print("current queue:") 
       for i, student in enumerate(queue, start=1):
           print(f"{i}. {student}")
```
